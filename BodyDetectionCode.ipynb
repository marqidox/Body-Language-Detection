## Read in collected data and process

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("training_coordinates2.csv")
```

```python
# x1, y1, z1... are features
# class is the target
x = df.drop("class", axis=1) # features
y = df['class'] # target value
```

```python
x_train, x_test, y_train, y_test =  train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train.values # dataframe
```

## Train ML Classification Model

```python
# make a ml pipeline, append custom functions to model
from sklearn.pipeline import make_pipeline 
# normalizing the dataset, to avoid overrepresentation of certain feature
from sklearn.preprocessing import StandardScaler

# use 4 classification algorithms and pick the best one
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
pipelines = {
    'lr':make_pipeline(StandardScaler(), LogisticRegression()),
    'rc':make_pipeline(StandardScaler(), RidgeClassifier()),
    'rfc':make_pipeline(StandardScaler(), RandomForestClassifier()),
    'gbc':make_pipeline(StandardScaler(), GradientBoostingClassifier()),
}
```

```python
list(pipelines.values())[0]
```

```python
# warning explaination: model convergence (Δerror is very small per iteration)
# can increase max iter for ml algos (default=100)
# https://stackoverflow.com/questions/62658215/convergencewarning-lbfgs-failed-to-converge-status-1-stop-total-no-of-iter
fit_models = {}
for algorithm, pipeline in pipelines.items():
    model = pipeline.fit(x_train.values, y_train) # training model, changed to values to avoid feature name error
    fit_models[algorithm] = model
```

```python
fit_models
```

## Evaluate and Serialize (save) Model

```python
fit_models['rfc'].predict(x_test.values) # pass through row of data to make real-time detections
```

```python
from sklearn.metrics import accuracy_score # accuracy metrics, just one metric
import pickle # save the models to the disk
```

```python
# normally try out in different situations to make sure models are accurate across the board
for algorithm, model in fit_models.items():
    yhat = model.predict(x_test) # testing models
    print(f"{algorithm} Accuracy: {accuracy_score(y_test, yhat)}")
```

# Creating the messages

```python
class Student:
    def __init__(self):
        self.detected_emotion = None

    def update_emotion(self, s_class):
        self.detected_emotion = s_class
        return self.send_message_to_student()

    def get_emotion(self):
        return self.detected_emotion

    def send_message_to_student(self): # works
        # could make a bank of messages and select randomly from it
        if self.detected_emotion == "tired":
            return "Engage in class some more!"
        elif self.detected_emotion == "sleepy":
            return "Stay awake by being active! Or, ask for a break."
        elif self.detected_emotion == "confused":
            return "Raise your hand to ask the teacher for clarification."
        else:
            return
```
